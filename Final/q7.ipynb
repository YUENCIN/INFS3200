```python
df1 = pd.read_csv("Book1.csv",names=['ID','Title','Authors','Publication_year','Publication_month','Publication_day','Edition','Publisher_name','ISBN13','Language','Series','Pages'])
df2 = pd.read_csv("Book2.csv",names=['ID','Title','Authors','Publication_year','Publication_month','Publication_day','Edition','Publisher_name','ISBN13','Language','Series','Pages'])
```

```python
idx1 = np.array(df1['ID'])
idx2 = np.array(df2['ID'])
title1 = np.array(df1['Title'])
title2 = np.array(df2['Title'])
```

```python
thresh_jd = 0.75
q = 3
results = []
for i in range(0,len(title1)):
    title_1 = str(title1[i])
    if (i%50==0):
        print(i/len(title1)*100)
    for j in range(i+1,len(title2)):
        title_2 = str(title2[j])
        jd = similarity.calc_jaccard(title_1,title_2,q)
        if jd >thresh_jd:
            results.append([idx1[i],idx2[j]])
```

```python
def load_benchmark():
    try:
        with open('Book1and2_pair.csv', 'r') as f:
            benchmark = f.read().splitlines()
        if benchmark is None:
            print("no file.")
        return benchmark
        # print(benchmark)

    except:
        print("Error occurred. Check if file exists")
def calc_measure(results):
    benchmark = load_benchmark()
    if len(results) == 0:
        print('Precision = 0, Recall = 0, Fmeasure = 0')
        return
    count = 0
    for pair in results:
        if pair in benchmark:
            count = count + 1
    if count == 0:
        print('Precision=0, Recall=0, Fmeasure=0')
        return
    precision = count / len(results)
    recall = count / len(benchmark)
    f_measure = 2 * precision * recall / (precision + recall)
    print("Precision=", precision, ", Recall=", recall, ", Fmeasure=", f_measure)
    return
```

```python
result = []
for v in results:
    result.append(str(v[0])+','+str(v[1]))
result
```

```python
calc_measure(result)
```
